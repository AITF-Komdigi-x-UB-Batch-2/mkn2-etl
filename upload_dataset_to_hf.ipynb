```python
from huggingface_hub import HfApi

api = HfApi()
```

```python
repo_id = "lfixfey/mkn_vlm_dataset"

folders_to_upload = ["rlh_int", "rlh_ext"]

for folder_name in folders_to_upload:
    print(f"Sedang mengunggah {folder_name}...")
    api.upload_folder(
        folder_path=f"data/mkn_img/{folder_name}",   # Path lokal ke sub-folder
        path_in_repo=f"data/mkn_img/{folder_name}",  # Path tujuan di HF (tetap di dalam mkn_img)
        repo_id=repo_id,
        repo_type="dataset",
        commit_message=f"Overwrite folder {folder_name} with updated data"
    )

print("Proses upload selesai!")
```

```python
api.upload_file(
    path_or_fileobj="data/cnn/mkn_image_metadata.json", # Path lokal file
    path_in_repo="data/mkn_img/mkn_image_metadata.json", # Path tujuan di repo HF
    repo_id=repo_id,
    repo_type="dataset",
    commit_message="Upload image level metadata to mkn_img folder"
)
```
